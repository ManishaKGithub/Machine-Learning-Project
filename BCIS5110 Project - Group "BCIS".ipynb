We will now delete all records with the status 'fundraising' from our database because they are no longer relevant to our next steps. Because we want to make the modifications to our dataset permanently, we'll use the drop command to ignore the fundraising data by setting it to True. We'll acquire the dataset that doesn't have fundraising status in our loans dataset when we finish the first stage.

We will replace the records of "expired," "refunded," and "notfunded" with "notfunded" because both records transmit the same record facts. We will also change the record name to a common one so that we can analyze the data more thoroughly.

```python
#'funded' suggests the loan is funded
#expired and refunded suggests the loan is not funded 
#fundraising can be ignored.

#deleting the fundraising value from 'STATUS'. we use filtering.
completed_Loans= loansdf['STATUS'] != 'fundRaising'
loansdf= loansdf[completed_Loans]

#recoding the categories into two categories
loansdf['STATUS']=loansdf['STATUS'].replace('expired','notfunded')
loansdf['STATUS']=loansdf['STATUS'].replace('refunded','notfunded')

#recodedStatus is target variable.
recodedStatus=loansdf['STATUS'].unique()
print(recodedStatus)

#checking the table
loansdf.head()
```

Examining the variable 'LOAN_AMOUNT'. <br>
1) Checking the distribution. What is the value of outliers if we use 3*IQR to define? <br>
2) Removing the outliers. <br>
3) Creating a new variable by taking log of this variable.

The interquartile range concept will be utilized now, for analyzing the distribution and the dispersion of data, we will use the IQR concept and will split the data as required with respect to quartiles. The observations which fall outside the mentioned quartiles are removed. We will remove the outliers by dropping the data which is above the upper range and below the lower range. The log function is utilized for creating a new variable by taking the log of the ‘LOAN AMOUNT’.

```python
#describe the 'LOAN_AMOUNT' to check the distribution
loansdf['LOAN_AMOUNT'].describe()
Q3=np.quantile(loansdf['LOAN_AMOUNT'] , 0.75)
Q1=np.quantile(loansdf['LOAN_AMOUNT'] , 0.25)

#IQR 
IQR = (Q3 - Q1)
print(IQR)

lowerVal = (Q1 - (3 * IQR))
upperVal = (Q3 + (3 * IQR))

#remove the outliers
loansdf.drop(loansdf[ (loansdf['LOAN_AMOUNT'] > upperVal) | (loansdf['LOAN_AMOUNT'] < lowerVal) ].index ,
             inplace=True)
#taking log and creating new variable
loansdf['LOAN_AMOUNT_LOG']=np.log(loansdf['LOAN_AMOUNT'])
loansdf['LOAN_AMOUNT_LOG'].value_counts()
loansdf.head()
```

Creating a new variable 'TIME_LENGTH' that is the difference between 'POSTED_TIME' and 'RAISED_TIME'.

A new variable called 'TIME_LENGTH' is created which is the difference between 'RAISED_TIME' and 'POSTED_TIME' is created with the help of pd.to_datetime function, the function utilizes the date and time on which the data is created.

```python
#creating a new variable 'TIME_LENGTH' which is the difference between RAISED_TIME and POSTED_TIME.
loansdf['TIME_LENGTH'] = pd.to_datetime(loansdf['RAISED_TIME']) - pd.to_datetime(loansdf['POSTED_TIME'])
loansdf['TIME_LENGTH'].dtype
loansdf['TIME_LENGTH']
#checking if the new variable is added in the table
loansdf.head()
```

Finding the difference between 'POSTED_TIME' and 'DISBURSE_TIME'. Creating a new variable 'PREDISBURSE' if the 'DISBURSE_TIME' is before 'POSTED_TIME'.

'POSTED_TIME' and 'DISBURSE_TIME' are the variables which describes the posted time details and disburse time details with respect to datetime function of panda. The 'PREDISBURSE' variable is 
created with respect to 'POSTED_TIME' and 'DISBURSE_TIME'. The 'PREDISBURSE' variable has the values where the 'DISBURSE_TIME' values less than 'POSTED_TIME' values.

```python
#creating a new variable 'PREDISBURSE' if the 'DISBURSE_TIME' is before 'POSTED_TIME'. Filling it with 1 else with 0.
loansdf.loc[pd.to_datetime(loansdf['DISBURSE_TIME']) < pd.to_datetime(loansdf['POSTED_TIME']), 'PREDISBURSE' ]=1
loansdf.loc[pd.to_datetime(loansdf['DISBURSE_TIME']) >= pd.to_datetime(loansdf['POSTED_TIME']), 'PREDISBURSE' ]=0

#checking the values for 1 and 0 category of PREDISBURSE variable.
loansdf['PREDISBURSE'].value_counts()

#checking the table
loansdf.head()                                                                   
```

Examining variable 'CURRENCY_EXCHANGE_COVERAGE_RATE'. Checking the unique values for this variable and the missing value count?<br>
Before we rush to replace missing values with mode, we ask the question: is the missing value actually missing or is it something else? We then read the policies on Kiva about exchange coverage. For instance, a lender gives 25 in US dollars and it changes to 250 in another currency. But when the money is repaid at 250 in the other currency, it only exchanges to 20 US dollars. The CURRENCY_EXCHANGE_COVERAGE_RATE measures how much such loss will be covered by the field partners. Some would cover 10% or 20%. So, it is more likely that the missing values indicating the loan does not have any coverage. So, it's more likely means 0.<br>
Replacing the missing value with 0.

To analyze the CURRENCY_EXCHANGE_COVERAGE_RATE', the missing values and unique values are found out from the dataset. The CURRENCY_EXCHANGE_COVERAGE_RATE variable measures how much loss will be covered by the field partners during the currency conversion. Some conversion would result in percentages such as 10% or 20%. So, it is indicated that the missing values indicating the loan does not have any coverage. So, it's more likely means 0. The missing values are replaced with 0 so that the currency conversion rates which will be having least effect on the dataset.

```python
#describing the variable 'CURRENCY_EXCHANGE_COVERAGE_RATE'
loansdf['CURRENCY_EXCHANGE_COVERAGE_RATE'].describe()

#finding the unique values
loansdf['CURRENCY_EXCHANGE_COVERAGE_RATE'].unique()

#finding the number of missing values
loansdf['CURRENCY_EXCHANGE_COVERAGE_RATE'].isnull().sum()

#filling the missing values with 0
loansdf['CURRENCY_EXCHANGE_COVERAGE_RATE'].fillna(0, inplace =True)

#checking the unique values to see if 0 is replaced in the missing values.
loansdf['CURRENCY_EXCHANGE_COVERAGE_RATE'].unique()
loansdf['CURRENCY_EXCHANGE_COVERAGE_RATE'].value_counts()
```

Now we can drop all records with missing values at this point.

```python
#At this point, we can drop all the records having missing values for better model predictions
loansAnalysisDf = loansAnalysisDf.dropna()
```

```python
#Here we are considering our target variable as 'STATUS'
targetVariable = loansAnalysisDf['STATUS']

#Below are the predictor variables we will be using.
predictFeatures = ['ORIGINAL_LANGUAGE','LOAN_AMOUNT_LOG','IMAGE_ID','SECTOR_NAME','COUNTRY_NAME','CURRENCY_POLICY',
                   'CURRENCY_EXCHANGE_COVERAGE_RATE','Days','PREDISBURSE','LENDER_TERM','NUM_JOURNAL_ENTRIES',
                   'NUM_BULK_ENTRIES','REPAYMENT_INTERVAL','DISTRIBUTION_MODEL']
Z = loansAnalysisDf[predictFeatures] 
Z.head()
```

Generating dummies for all categorical variables.

```python
#we take the below as categorical variables
categoricalVar = ['ORIGINAL_LANGUAGE','SECTOR_NAME','COUNTRY_NAME','CURRENCY_POLICY','PREDISBURSE',
                  'REPAYMENT_INTERVAL','DISTRIBUTION_MODEL']

#generating dummies for all our categorical variables
Z = pd.get_dummies(Z , columns=categoricalVar, drop_first = True)

#checking the dummy values in the table
Z.head()
```

Preparing training and test datasets.

* X_test - This is the remainder of the data's independent variables that will not be included in the training phase but will be used to make predictions to assess the model's accuracy.<br>
* X_train - This is the training dataset<br>
* y_test - This data has category labels for your test data, these labels will be used to test the accuracy between actual and predicted categories.<br>
* y_train - this is the dependent variable which needs to be predicted by this model, this includes category labels against your independent variables, we need to specify our dependent variable while training/fitting the model.
<br>

It is utilized to address the overfitting issues present in the model<br>
We will be importing the train_test_split from sklearn.model_selection

```python
#Preparing training and test datasets
#importing train_test_split from sklearn package.
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(Z, targetVariable, test_size = 0.33,random_state =1)
```

Fit a logistic regression model.<br>
